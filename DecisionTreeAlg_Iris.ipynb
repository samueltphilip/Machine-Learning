```python
import numpy as np
import pandas as pd
df=pd.read_csv("/home/samuel/Downloads/Luminar/analysis_files/Iris.csv")
df.head()
```

```python
# reading values
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
# splitting into train and test sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
# decision tree classifier
from sklearn import tree # OR from sklearn.tree import DecisionTreeClassifier
dt=tree.DecisionTreeClassifier() 
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
y_pred
```

```python
# see the tree structure built by the machine
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))
tree.plot_tree(dt,feature_names=['sepal_length','sepal_width','petal_length','petal_width'],class_names=['Iris-versicolor','Iris-virginica','Iris-setosa'],filled=True)# fill=true is to see colour
```

```python
# performance evaluation
from sklearn.metrics import classification_report,confusion_matrix
print("Classification Report",classification_report(y_test,y_pred))
print("Classification Report",confusion_matrix(y_test,y_pred))
```
