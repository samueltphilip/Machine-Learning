```python
# From customer file, fill missing values and find each location/profession people count

# importing libraries and file to be analysed
import numpy as np #numerical python
import pandas as pd # data manipulation,cleaning and analysis
import matplotlib.pyplot as plt #data visualization
cust=pd.read_csv("/home/samuel/Downloads/analysis_files/customer",header=None) # read file
cust.columns=['id','fname','lname','age','profession','location'] #assigning column names
print(cust.head())
```

```python
# looking for missing values
print(cust.isna().sum())
cust['location'].fillna('india', inplace=True) #filling missing value
```

```python
print(cust.isna().sum())
```

```python
# grouping by location
cus=cust.groupby('location')['location'].count()
print((cus))
```

```python
print(cus.index)
```

```python
# visualization (pie chart)
plt.pie(cus,labels=['africa', 'australia', 'china', 'india', 'ireland', 'uk', 'us'])
plt.legend(bbox_to_anchor=(3,1)) # placing colour legend
plt.show()
```

```python
# grouping by profession
cuso=cust.groupby('profession')['profession'].count()
print((cuso))
```

```python
print(cuso.index)
```
